#Variable Declarations and Base Model name assignments

```python
# Importing the Libraries
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig,HfArgumentParser,TrainingArguments,pipeline, logging, TextStreamer
from peft import LoraConfig, PeftModel, prepare_model_for_kbit_training, get_peft_model
import os, torch, wandb, platform, warnings
from datasets import load_dataset
from trl import SFTTrainer
from huggingface_hub import notebook_login
#Use a sharded model to fine-tune in the free version of Google Colab.
#Instead of loading the entire model into memory, sharding allows loading and processing only the necessary parts,
# reducing memory requirements significantly. Faster Inference: By distributing the computation across multiple devices,
# sharding helps achieve parallelism, resulting in faster inference times
base_model = "mistralai/Mistral-7B-v0.1" #bn22/Mistral-7B-Instruct-v0.1-sharded
```

#Mistral Model's Specific Data Template adopted to form the Harry Potter's Conversational Dataset

```python
#Loading the Harry Potter Dialog Dataset formated as per the Mistral Model's requirement
#Importing required libraries such as
#Pandas for CSV file reading. Dataset for transforming Pandas Dataframe into a hugging face dataset
#randrange for randomly selecting to review the data
import pandas as pd
from datasets import Dataset
from random import randrange
# Loading dataset from CSV file. it contains three columns namely: instruction, input and output
#formated using the jupyter notebook ChangeDataformatForLlama2_mistral_HPCSVw9.jpynb #HP_dialogdataset_Mistral.csv
df = pd.read_csv("HP_dialogdataset_Mistral.csv")
# Converting the DataFrame to Data Dictionary to match the Hugging Face Dataset format
dataset_dict = {
    "chat_sample": df["chat_sample"].tolist(),
    "dataset_origin": df["dataset_origin"].tolist(),
}
# Creating Dataset object
dataset = Dataset.from_dict(dataset_dict)

print(f"Dataset Size: {len(dataset)}")
print(dataset[randrange(len(dataset))])
```

#Instantiated the base model("mistralai/Mistral-7B-v0.1") using 4-bit quantization using BitsAndBytesConfig() method. And Instantiated the base model's  tokenizer
The packages bitsandbytes, 4-bit quantization and LoRA are used here to make the MISTAL-7B kind of LLM  even more accessible for fine-tuning despite of its large size due to more parameters. Below code does the Quantization of the large   MISTAL-7B kind pre-trained model to 4 bits and freeze it. This will attach small, trainable adapter layers (named as LoRA). Finally, fine-tuning only the adapter layers while utilizing the frozen quantized model for context.


```python
# Loading the Mistral 7B as the base model
bnb_config = BitsAndBytesConfig(
    load_in_4bit= True,
    bnb_4bit_quant_type= "nf4",
    bnb_4bit_compute_dtype= torch.bfloat16,
    bnb_4bit_use_double_quant= False
)
model = AutoModelForCausalLM.from_pretrained(
    base_model,
    quantization_config=bnb_config,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    trust_remote_code=True
)
model.config.use_cache = False # This setting will silence the warnings. But it should be re-enabled for inference!
model.config.pretraining_tp = 1
model.gradient_checkpointing_enable()
# Loading the Mistral-7B Model's tokenizer
tokenizer = AutoTokenizer.from_pretrained(base_model, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.add_eos_token = True
tokenizer.add_bos_token, tokenizer.add_eos_token
#RuntimeError: No GPU found. A GPU is needed for quantization. This ensured that MSITRAL-7B fine-tuning is possible only with GPU and not with CPU.
```

#Prepare and configure the Model for knowledge distillation and the LoRA technique for efficient training
INSTANTIATING THE LoRA CONFIGURATION BASED ON THE QLoRA PAPER

```python
model = prepare_model_for_kbit_training(model)
peft_config = LoraConfig(
        r=16,
        lora_alpha=16,
        lora_dropout=0.05,
        bias="none",
        task_type="CAUSAL_LM",
        target_modules=["q_proj", "k_proj", "v_proj", "o_proj","gate_proj"]
    )
model = get_peft_model(model, peft_config)
```

###"SFTTrainer" was used for training the model instead of "Trainer".
SFTTrainer could be used on a pre-trained model having a relatively smaller dataset and wants a simpler and faster fine-tuning experience with efficient memory usage. It could be noted that the SFTTrainer offers seamless integration with peft, simplifying the process of efficiently fine-tuning LLMs with instructions. Hence, in the below code the LoRa congiruration is created as "LoRAConfig" and supplyed it to the trainer.
#Setting the hyperparameters as the trainer arguments and the "SFTTrainer"  for finetuning the Mistral-7B
## Mistral Model (hpv2_finetuned-Mistral-7B) Training done with 2 Epcohs and 1000 save_steps

```python
#Mistral Model (hpv2_finetuned-Mistral-7B) Training done with 2 Epcohs and 1000 save_steps
#Setting the Training Arguments before instantiating the Model training #from transformers import TrainingArguments, SFTTrainer
# Hyperparameters are adjusted based on the hardware usage
training_arguments = TrainingArguments(
    output_dir= "hpv2_finetuned-Mistral-7B",  # Final Fine-tuned model
    logging_dir="logs",                      #Logs stored under logs directory
    num_train_epochs= 2,                     #Number of training epochs
    per_device_train_batch_size= 8,          #batch size per training device
    gradient_accumulation_steps= 2,          #number of gradient accumulation steps
    optim = "paged_adamw_8bit",              #"paged_adamw_8bit" optimizer used for training
    save_steps= 1000,                        #frequency of saving checkpoints during training.
    logging_steps= 30,                       # 30 is the frequency of logging training metrics.
    learning_rate= 2e-4,                     #initial learning rate for the optimizer.
    weight_decay= 0.001,                     #weight decay parameter for regularization
    fp16= False,                             # not using  mixed-precision training with float16 for reduced memory usage
    bf16= False,                             #  not using mixed-precision training with bfloat16
    max_grad_norm= 0.3,                      # maximum gradient norm for gradient clipping.
    max_steps= -1,                           # maximum number of training steps
    warmup_ratio= 0.3,                       # ratio of warmup steps during learning rate warmup
    group_by_length= True,                   #group samples by length for more efficient training.
    lr_scheduler_type= "constant",           # type of learning rate scheduler used during training.
    push_to_hub = True,                      # pushed the fine-tuned model to HuggingFace model hub
    report_to=["tensorboard", "wandb"],      # the platforms to report training metrics to (e.g., TensorBoard and Weights & Biases).
)
# Setting sft parameters
trainer = SFTTrainer(
    model=model,              # Base Model
    train_dataset=dataset,    # Conversational Dataset
    peft_config=peft_config,  # assigning the PEFT configuration for quantization
    max_seq_length= None,    # Maximum sequence length is no limited here by setting to None
    dataset_text_field="chat_sample",    #Particular Field/Column name "chat_sample" of the Dialogs column in the Conversational Dataset
    tokenizer=tokenizer,    #Base model's tokenizer
    args=training_arguments,    #setting the training-arguments
    packing= False,    #Packing not required
)
```

#Training the Mistral-7B model using the "SFTTrainer" trainer on the Conversational dialog dataset and saving the copy of trained model in the local drive

```python
# Training the Mistral model using the "SFTTrainer" trainer on the Harry potter dialog dataset and saving the model in local disk
trainer.train()

# Saving the trained(fine-tuned) model in local disk
trainer.save_model()
```
